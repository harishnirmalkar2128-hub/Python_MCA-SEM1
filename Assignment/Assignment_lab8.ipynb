```python
# Sample input: list of files
files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']

# Dictionary to store files by extension
files_by_ext = {}

for file in files:
    # Split the filename and extension
    name, ext = file.rsplit('.', 1)
    ext = ext.lower() # normalize extension

    # Add file to the directory
    if ext not in files_by_ext:
        files_by_ext[ext] = []
    files_by_ext[ext].append(file)

# Display the categorized dictionary
print(files_by_ext)
```
