```python
import string

text = "Rain rain go away, come again another day."

# Convert to lowecase
text = text.lower()

# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# Split into words
words = text.split()

# Count frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```
