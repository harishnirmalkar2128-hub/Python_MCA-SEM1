```python
# Program to count lines, words, and characters in a text file

def count_file_stats(filename):
    with open(filename, "r") as file:
        text = file.read()

    # Counting lines
    with open(filename, "r") as file:
        lines = file.readlines()
    num_lines = len(lines)

    # Counting Words
    words = text.split()
    num_words = len(words)

    # Counting characters (including spaces and newline characters)
    num_char = len(text)

    return num_lines, num_words, num_char

# Example usage
filename = "sample.txt"
lines, words, chars = count_file_stats(filename)

print("Lines:", lines)
print("Words:", words)
print("Characters:", chars)
```
