```python
# List of employee records (each record is a dictionary)
employees = [
    {"name": "Nidhi", "salary": 45000},
    {"name": "Harish", "salary": 32000},
    {"name": "Chaitanya", "salary": 55000},
    {"name": "Sameer", "salary": 38000}
]

# Sorting the employees list by salary
sorted_employees = sorted(employees, key=lambda emp: emp["salary"])

# Displaying the Sorted list
print("Employees sorted by salary:")
for emp in sorted_employees:
    print(emp)
```
