Pattern Programs

```python
print("Number pattern \n")
Number=6
for col in range(1, Number):
    for row in range(1, col + 1):
        print(row, end=" ")
    print()  #new line
```

```python
print("Pyramid Pattern \n")
rows = 5
for i in range (1, rows + 1):
    print(" " * (rows - i) + "*" * (2*i -1))
```

```python
print("Diamond Pattern \n")

row_cols=5
# For a diamond with maximum width of 5 stars
# The pattern should have: 1, 3, 5, 3, 1 stars
    
# Upper half
for i in range(1, row_cols + 1, 2):
        spaces = (row_cols - i) // 2
        print("  " * spaces + "* " * i)
    
# Lower half
for i in range(row_cols - 2, 0, -2):
        spaces = (row_cols - i) // 2
        print("  " * spaces + "* " * i)
```

```python
print("Hollow square pattern \n")

Number = 4
for col in range(Number):
    for row in range(Number-1):
        if col == 0 or col == Number - 1 or row == 0 or row == Number - 1 - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
print("Arrow-like Pattern \n")
rows = 5

# Upper arrow
for i in range (1, rows + 1):
    print(" " * (rows - i) + "* " * i)
```
