```python
# Prpgram to search for an item in a list

# Sample Input 
num_list = [12, 45, 33, 21, 88]
item = 33

# Search for the item
if item in num_list:
    position = num_list.index(item) + 1 # +1 to make position human - readable
    print(f"Item found at position {position}")
else:
    print("Item not found.")
```
