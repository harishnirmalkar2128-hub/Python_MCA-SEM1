```python
import csv
from collections import defaultdict

# Dictionary to store salaries by department
dept_salaries = defaultdict(list)

# Open and read the CSV file
with open('employees.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        name, department, salary = row
        dept_salaries[department].append(float(salary))

# Calculate and display average salary by department
print("Average Salary by Department:")
for department, salaries in dept_salaries.items():
    average_salary = sum(salaries) / len(salaries)
    print(f"{department}: {average_salary}")
```
