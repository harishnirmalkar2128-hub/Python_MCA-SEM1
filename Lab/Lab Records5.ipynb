```python
def reverse_vowels(s):
    vowels = "aeiouAEIOU"
    s_list = list(s)
    
    # Collect vowels in the string
    vowel_positions = [i for i, char in enumerate(s_list) if char in vowels]
    vowel_chars = [s_list[i] for i in vowel_positions]
    
    # Reverse the vowels
    vowel_chars.reverse()
    
    # Place reversed vowels back
    for i, pos in enumerate(vowel_positions):
        s_list[pos] = vowel_chars[i]
    
    return "".join(s_list)


# Input
sentence = input("Enter sentence: ")

# Output
reversed_vowels_sentence = reverse_vowels(sentence)
print(reversed_vowels_sentence)
```
