```python
import random
import string

# Define character pools
uppercase = string.ascii_uppercase
lowercase = string.ascii_lowercase
digits = string.digits
specials = "!@#$*"

# First character: Capital letter or digit
first_char = random.choice(uppercase + digits)

# Remaining 11 characters: mix of all allowed characters
all_chars = uppercase + lowercase + digits + specials
remaining_chars = [random.choice(all_chars) for _ in range(11)]

# Combine first character with the rest
password = first_char + "".join(remaining_chars)

print("Random Password:", password)
```
