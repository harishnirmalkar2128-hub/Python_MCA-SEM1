```python
# Input
num = int(input("Enter a number: "))

# Convert to different bases
binary = bin(num)[2:]       # Remove '0b' prefix
octal = oct(num)[2:]        # Remove '0o' prefix
hexadecimal = hex(num)[2:]  # Remove '0x' prefix

# Display results
print("Binary:", binary)
print("Octal:", octal)
print("Hexadecimal:", hexadecimal.upper())
```
