```python
import math

# Input number for square root and factorial
num = int(input("Enter a number: "))

# Calculate square root
sqrt_num = math.sqrt(num)
print("Square root of", num, "is", sqrt_num)

# Calculate factorial
fact_num = math.factorial(num)
print("Factorial of", num, "is", fact_num)

# Input base and exponent for power
base = float(input("Enter the base for power calculation: "))
exponent = float(input("Enter the exponent: "))

# Calculate power
power_num = math.pow(base, exponent)
print(f"{base} raised to the power {exponent} is {power_num}")
```
